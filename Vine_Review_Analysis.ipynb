Load Data From AWS

```python
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Video_Games_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get(""), sep="\t", header=True, inferSchema=True)
df.show()
```

```python
from pyspark.sql.functions import to_date
vine_df = df.select(["review_id", "star_rating", "helpful_votes", "total_votes", "vine", "verified_purchase"])
vine_df.show(10)
```

```python
vine_filtered_df = vine_df.filter("total_votes >= 20")
vine_filtered_df.show(10)
```

```python
#Filter for more than %50 helpful votes.
vine_helpful_votes_df = vine_filtered_df.filter("(helpful_votes/total_votes) >= .5")
vine_helpful_votes_df.show(10)
```

```python
#Filter for paid reviews.
vine_helpful_paid_df = vine_helpful_votes_df.filter("vine == 'Y'")
vine_helpful_paid_df.show(10)
```

```python
vine_helpful_unpaid_df = vine_helpful_votes_df.filter("vine == 'N'")
vine_helpful_unpaid_df.show(10)
```

```python
total_reviews = vine_helpful_votes_df.count()
total_reviews
```

```python
#Total paid reviews.
total_reviews_paid_df = vine_helpful_votes_df.filter("vine == 'Y'")
total_reviews_paid_count = total_reviews_paid_df.count()
total_reviews_paid_count
```

```python
#Total unpaid reviews
total_reviews_unpaid_df = vine_helpful_votes_df.filter("vine == 'N'")
total_reviews_unpaid_count = total_reviews_unpaid_df.count()
total_reviews_unpaid_count
```

```python
#5 star reviews.
total_five_star_df = vine_helpful_votes_df.filter("star_rating == 5")
total_five_star_count = total_five_star_df.count()
total_five_star_count
```

```python
#5 star paid reviews.
five_star_paid_df = total_reviews_paid_df.filter("star_rating == 5")
five_star_paid_count = five_star_paid_df.count()
five_star_paid_count
```

```python
#5 star unpaid reviews.
five_star_unpaid_df = total_reviews_unpaid_df.filter("star_rating == 5")
five_star_unpaid_count = five_star_unpaid_df.count()
five_star_unpaid_count
```

```python
#5 star paid review percent.
five_star_paid_percent = (five_star_paid_count / total_five_star_count)*100
five_star_paid_percent
```

```python
#5 star unpaid review percent.
five_star_unpaid_percent = (five_star_unpaid_count / total_five_star_count) * 100
five_star_unpaid_percent
```
